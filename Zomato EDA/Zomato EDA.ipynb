```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('zomato.csv', encoding='latin-1')
data
```

#### There is a column named 'Country Code'. We need to replace it with country name. Kaggle has provided another file with country codes and names in it.

```python
count_code = pd.read_excel('CountCode.xlsx')
count_code
```

```python
d1 = data.copy()
```

```python
final_data = d1.merge(count_code, on='Country Code', how='left')
final_data
```

```python
final_data.drop('Country Code', axis='columns', inplace=True)
```

```python
final_data
```

```python
# Checking out for null values

final_data.isnull().sum()
```

```python
final_data[final_data['Cuisines'].isnull()]
```

We can see that all the records with missing cuisins are from United States

```python
final_data.dropna(inplace=True)
final_data.isna().sum()
```

```python
final_data.info()
```

```python
final_data.describe()
```

```python
plt.figure(figsize=(12, 7))
ax = sns.heatmap(final_data.corr(), annot=True, linewidth=2)
plt.tight_layout()
plt.show()
```

```python
country_wise_grouped = final_data.groupby('Country')
country_wise_grouped.get_group('India')
```

```python
final_data['Country'].unique()
```

```python
print(len(country_wise_grouped.get_group('India'))/len(final_data))
```

```python
zomato_operational_country = final_data['Country'].unique()
country_wise_exposure = pd.DataFrame({'Country':zomato_operational_country, '% Exposure':[(len(country_wise_grouped.get_group(i))/len(final_data))*100 for i in zomato_operational_country]})
```

```python
country_wise_exposure
```

```python
top_3 = country_wise_exposure.nlargest(3, '% Exposure')

plt.figure(figsize=(12, 7))
ax = plt.pie(top_3['% Exposure'], labels=top_3['Country'], autopct='%1.1f%%')
plt.title('Top 3 Countries', fontweight='bold')
plt.tight_layout()
plt.show()
```

##### Popular cuisines in India

```python
d1 = indian_exposure.copy()
d1['Cuisines'] = d1['Cuisines'].str.split(',')
d1
```

```python
indian_cuisines = []
for i in d1['Cuisines']:
    for j in i:
        if j.strip() not in indian_cuisines:
            indian_cuisines.append(j.strip())
            
indian_cuisines
```

```python
len(indian_cuisines)
```

##### Popularity of each cuisine

```python
all_cuisines = [cuisine.strip() for i in d1["Cuisines"] for cuisine in i]

cuisine_counts = pd.Series(all_cuisines).value_counts()

plt.figure(figsize=(15, 6))
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Popularity of Cuisines')
plt.tight_layout()
plt.show()
```

```python
cuisine_counts = cuisine_counts[cuisine_counts.values>30]

plt.figure(figsize=(15, 8))
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Popularity of Cuisines in india')
plt.tight_layout()
plt.show()
```
